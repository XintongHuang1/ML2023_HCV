# Histogram of each numeric feature

```python
fig,ax=plt.subplots(3,4,figsize=(15,10))
ax[0][0].hist(data['Age'])
ax[0][0].set_title('Age')
ax[0][1].hist(data['ALB'])
ax[0][1].set_title('ALB')
ax[0][2].hist(data['ALP'])
ax[0][2].set_title('ALP')
ax[0][3].hist(data['ALT'])
ax[0][3].set_title('ALT')
ax[1][0].hist(data['AST'])
ax[1][0].set_title('AST')
ax[1][1].hist(data['BIL'])
ax[1][1].set_title('BIL')
ax[1][2].hist(data['CHE'])
ax[1][2].set_title('CHE')
ax[1][3].hist(data['CHOL'])
ax[1][3].set_title('CHOL')
ax[2][0].hist(data['CREA'])
ax[2][0].set_title('CREA')
ax[2][1].hist(data['GGT'])
ax[2][1].set_title('GGT')
ax[2][2].hist(data['PROT'])
ax[2][2].set_title('PROT')
ax[2][3].hist(data['Sex'])
ax[2][3].set_title('Sex')
plt.show()
```

# 11 variables boxplot

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.set_title('11 Variables Boxplot')

df = data.drop('Category', axis = 1)
df = df.drop('Sex', axis = 1)

ax.boxplot(df, labels=df.columns)

plt.show()
```

1. Category (categorical)

```python
category = np.unique(data['Category'],return_counts=True)[0]
count = np.unique(data['Category'],return_counts=True)[1]
fig, ax = plt.subplots(figsize=(12,8))
ax = sns.barplot(x = category, y = count)
ax.set_xlabel('Category')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12,8))
ax.pie(count, labels = category)
ax.set_title('Category')
plt.show()
```

# Correlation matrix / heat map

```python
matrix = data.corr()
cmap = sns.diverging_palette(250, 15, s=75, l=40, n=9, center="light", as_cmap=True)
 
plt.figure(figsize=(12, 8)) 
sns.heatmap(matrix,  center=0, annot=True,
             fmt='.2f', square=True, cmap=cmap)
plt.show()
```
